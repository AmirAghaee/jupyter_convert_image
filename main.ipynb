```python
def convert_image(image_path, save_path):

    # 1. Opening the image:
    im = Image.open(image_path)
    # 2. Converting the image to RGB colour:
    im = im.convert('RGB')
    # 3. Spliting the image path (to avoid the .jpg or .png being part of the image name):
    image_name = image_path.split('.')[0]
    # Saving the images based upon their specific type:
    im.save(image_name + '.webp', 'webp')
    im.save(image_name + '.jp2', 'JPEG2000', quality_mode='dB', quality_layers=[41])
    
```

```python
directories = [x[0] for x in os.walk('img')]
for directory in directories:
    files = os.listdir(directory)
    for image in files:
        if image.endswith('jpg'):
            convert_image(directory + "\\" + image, directory)
        elif image.endswith('png'):
            convert_image(directory + "\\" + image, directory)
            
            
```
